```python
# Import dependencies
import pandas as pd
import os
import requests
import json
import urllib
```

```python
# Read in the csv
job_df = pd.read_csv("Resources/DataAnalyst.csv")
job_df.head()
```

```python
# Drop columns we do not need for the project
job_df = job_df.drop(columns=['Unnamed: 0', 'Job Description', 'Founded', 'Competitors', 'Easy Apply'])
job_df.head()
```

```python
job_df.describe()
```

```python
job_df.dropna()
```

```python
# Drop rows with -1 in them to see how many jobs left
# NOTE: Rating isn't dropping - needed to polish as an xlsx file
clean_job_df = job_df[job_df["Headquarters"] !="-1"]
clean_job_df = clean_job_df[clean_job_df["Industry"] !="-1"]
clean_job_df = clean_job_df[clean_job_df["Sector"] !="-1"]
clean_job_df = clean_job_df[clean_job_df["Revenue"] !="-1"]
clean_job_df = clean_job_df[clean_job_df["Salary Estimate"] !="-1"]
clean_job_df = clean_job_df[clean_job_df["Rating"] !="-1.0"]
clean_job_df.head(20)
```

```python
clean_job_df.describe()
```

```python
clean_job_df = clean_job_df.dropna()
clean_job_df.head()
```
